```python
%matplotlib inline

import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.linspace(0, 10, 50)
x = list(x) # Optional
y1, y2, y3, y4 = np.sin(x), np.cos(x), np.tan(x), np.exp(x)
data = [y1, y2, y3, y4]
labels = ['sin', 'cos', 'tan', 'exp']
```

```python
rows, cols = 2, 2
pos = [(x,y) for x in range(rows) for y in range(cols)]
```

```python
canvas, fig = plt.subplots(rows, cols, figsize=(10, 5))

for idx in range(0, rows*cols):
    n = pos[idx] 
    fig[n].plot(x, data[idx])
    fig[n].set(xlabel='x', ylabel='y', title=labels[idx])
    plt.tight_layout()
canvas.suptitle("Matplotlib Plot");
```
